```python
import pandas as pd

df = pd.read_excel('NYC_CPI.xlsx')
```

```python
df.head(15)
```

```python
df = df[10:]
df.reset_index(inplace=True, drop=True)
df.columns = df.iloc[0]
df = df[1:]
df.reset_index(inplace=True, drop=True)
df['Year'] = df['Year'].astype(int)
df['Year'] = df['Year'].astype(str)
df = df[df['Year'] != '2022']
df.drop(['Annual', 'HALF1', 'HALF2'], axis=1, inplace=True)
```

```python
df.head(15)
```

```python
df.dtypes
```

```python
year_list = list()

for year in df['Year']:
    for month in (df.columns)[1:]:
        year_list.append(month + '-' + year)
```

```python
cpi_list = list()

for i in df.values:
    for j in i:
        if j not in df['Year'].values:
            cpi_list.append(j)           
```

```python
df_cleaned = pd.DataFrame()
df_cleaned['Year'] = year_list
df_cleaned['CPI'] = cpi_list
df_cleaned.head(24)
```

```python
df_cleaned.to_csv('CPI.csv', index=False)
```
