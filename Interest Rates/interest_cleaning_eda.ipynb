```python
import pandas as pd

df = pd.read_csv('FEDFUNDS.csv')
```

```python
for i in df.values:
    print(i[0])
```

```python
month_dict = {'01':'Jan', '02':'Feb', '03':'Mar', '04':'Apr', '05':'May', '06':'Jun', '07':'Jul', '08':'Aug',
            '09':'Sep', '10':'Oct', '11':'Nov', '12':'Dec'}

year_list = ['2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']
month_list = list()
interest_list = list()
```

```python
for i in df.values:
    if i[0][0:4] in year_list:
        month_list.append(str(month_dict[i[0][5:7]] + '-' + i[0][0:4]))
        interest_list.append(i[1])
```

```python
df_cleaned = pd.DataFrame()
df_cleaned['Year'] = month_list
df_cleaned['Interest_Rate'] = interest_list
```

```python
df_cleaned.to_csv('interestrates_cleaned.csv', index=False)
```
