```python
import pandas as pd

df = pd.read_csv('Annual_Population_Estimates_for_New_York_State_and_Counties__Beginning_1970.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['FIPS Code'] = df['FIPS Code'].astype(str)
df['Year'] = df['Year'].astype(str)

df.dtypes
```

```python
df.describe(include='all')
```

```python
FIPS_dict = {'36061':'Manhattan', '36005':'Bronx', '36047':'Brooklyn', '36081':'Queens', '36085':'Staten Island'}
Year_list = ['2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']
```

```python
invalid_FIPS = list()

for i in df['FIPS Code']:
    if i not in FIPS_dict:
        invalid_FIPS.append(i)
```

```python
df = df[df['FIPS Code'].isin(invalid_FIPS) == False]
```

```python
invalid_year = list()

for i in df['Year']:
    if i not in Year_list:
        invalid_year.append(i)
```

```python
df = df[df['Year'].isin(invalid_year) == False]
```

```python
df.describe(include='all')
```

```python
df.drop_duplicates(subset=['FIPS Code', 'Geography', 'Year'], keep='last', inplace=True)
```

```python
df.describe(include='all')
```

```python
import numpy as np

for county in FIPS_dict:
    df['Geography'] = np.where(df['FIPS Code'] == county, FIPS_dict[county], df['Geography'])
```

```python
df.to_csv('popestimate_cleaned.csv', index=False)
```
